```python
import pandas as pd
from datetime import datetime
# Load the dataset from the session memory
df = pd.read_csv('/content/train.csv')


df.head()
```

```python
# Convert the non numeric to numeric
df['Mileage'] = df['Mileage'].astype(str).str.extract('([0-9.]+)').astype(float)
df['Engine'] = df['Engine'].astype(str).str.extract('([0-9.]+)').astype(float)
df['Power'] = df['Power'].astype(str).str.extract('([0-9.]+)').astype(float)
df['New_Price'] = df['New_Price'].astype(str).str.extract('([0-9.]+)').astype(float)
```

```python
# Checking the  missing values
print(df.isnull().sum())

 #Impute numeric with median, categorical with mode
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].median())
df['Engine'] = df['Engine'].fillna(df['Engine'].median())
df['Power'] = df['Power'].fillna(df['Power'].median())
df['Seats'] = df['Seats'].fillna(df['Seats'].mode()[0])

df = df.dropna(subset=['Price'])
```

I used the median and mode here for replacing the missed values, median is robust to outliers. For discrete values mode is better.

```python
# One hot encoding for Fuel_Type and Transmission
hotEncoding = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'], drop_first=True)
print(hotEncoding)
```

```python
from datetime import datetime
current_year = datetime.now().year
hotEncoding['Car_Age'] = current_year - hotEncoding['Year']
print(hotEncoding)
```

```python
# Select
select_df = hotEncoding[[  'Location', 'Car_Age', 'Price']]

# Filter :cars > 10 years
filter_df = select_df[select_df['Car_Age'] > 10]

# Rename a column
rename_df = filter_df.rename(columns={'Price': 'ChangedPrice'})

# Mutate (add a new column = price per age)
rename_df['Price_per_Year'] = rename_df['ChangedPrice'] / rename_df['Car_Age']

#  price descending
arrange_df = rename_df.sort_values(by='ChangedPrice', ascending=False)

# Group by location and summarize average price
summary = arrange_df.groupby('Location').agg({'ChangedPrice': ['mean', 'max'], 'Car_Age': 'mean'}).reset_index()
print(select_df,rename_df,filter_df,arrange_df,summary)
```
